# tableau de bord de données

## Brève description

+ **Crée un tableau de bord HTML statique simple avec les données X, y fournies pour aider les utilisateurs à voir ce qui se passe dans leurs données, les aide à prendre des décisions concernant les fonctionnalités et trouve le meilleur modèle d'apprentissage automatique « de base ».**


## Description plus longue

+ **La bibliothèque `data_dashboard` vous permet de créer un tableau de bord HTML visualisant non seulement les données et les relations entre les fonctionnalités, mais également de rechercher automatiquement le meilleur modèle compatible avec `Sklearn « de base »`.**



+ Le tableau de bord HTML créé contiendra 3 sous-pages que vous pourrez consulter :

  + Aperçu avec statistiques récapitulatives des données
  + Fonctionnalités où vous pouvez approfondir chaque fonctionnalité des données
  + Modèles affichant les résultats de recherche et les performances des modèles
