# Basic steps for developing a model with Pycaret

```python
import pycaret
```

```python
pycaret.__version__
```

```python
from pycaret.datasets import get_data
data1 = get_data(dataset = 'traffic')
```

```python
from pycaret.regression import *
```

```python
dataset = setup(data = data1, target = 'traffic_volume',
                session_id = 438)
```

```python
best = compare_models()
```

```python
tuned_best = tune_model(estimator = best)
```

```python
final_model = finalize_model(estimator = tuned_best) 
```

```python
save_model(final_model, "models/rf_base_traffic")
```

#### Predictions

```python
saved_model = load_model('models/rf_base_traffic')
```

```python
from sklearn.model_selection import train_test_split

# Diviser les données en ensembles d'entraînement (80%) et de test (20%)
train_data, test_data = train_test_split(data1, test_size=0.2, random_state=42)

# Utilisez train_data pour l'entraînement du modèle et test_data pour les prédictions
# Ici, nous allons utiliser test_data comme new_data pour les prédictions
new_data = test_data
```

```python
predictions = predict_model(saved_model, data=new_data)
predictions.head()
```
