```python
import klib
import plotly
import pandas as pd
```

```python
print(pd.__version__)
print(klib.__version__)
print(plotly.__version__)
```

```python
file_path = "NFL_DATASET.csv"
```

```python
df = pd.read_csv(file_path).iloc[:,10:]
```

```python
df.head()
```

```python
df_cleaned = klib.data_cleaning(df)
```

```python
df_cleaned.head(3)
```

```python
klib.missingval_plot(df_cleaned)
```

```python
klib.cat_plot(df_cleaned.iloc[:,10:28], figsize=(24, 15))
```

```python
klib.corr_plot(df, annot=False, figsize=(20,17))
```

```python
klib.corr_interactive_plot(df, annot=False, figsize=(20,17))
```
