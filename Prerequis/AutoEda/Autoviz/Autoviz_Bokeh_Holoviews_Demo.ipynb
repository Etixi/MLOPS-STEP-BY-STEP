```python
import pandas as pd
from autoviz.AutoViz_Class import AutoViz_Class
# create a class
AV = AutoViz_Class()
```

```python
# load data and save it as csv
df0 = pd.read_csv('https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/MASS/Boston.csv')
df0.to_csv('file1.csv')
df0.head()
```

# For bokeh charts and dashboards either use "bokeh" or "server" as chart_format in your input

```python
# upload the data
# read a csv file
filename = "/content/file1.csv"
sep = ","
dft = AV.AutoViz(
    filename="",
    sep=",",
    depVar="medv",
    dfte=df0,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="bokeh",
    max_rows_analyzed=2000,
    max_cols_analyzed=20,
)
```

## To save your plots as Dynamic HTML files with widgets and dynamic drop down controls, choose "html" as chart_format

```python
# upload the data
# read a csv file
filename = "/content/file1.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="medv",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="html",
    max_rows_analyzed=2000,
    max_cols_analyzed=20,
)
```
