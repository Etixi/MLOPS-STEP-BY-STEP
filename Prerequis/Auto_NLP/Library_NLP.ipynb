####**NeatText**

####**Problème**
+ Nettoyage des données textuelles non structurées
+ Réduire le bruit [caractères spéciaux, mots vides]
+ Réduire la répétition de l'utilisation du même code pour le prétraitement du texte

####**La solution**
+ convertir la solution déjà connue pour nettoyer le texte en un package réutilisable


####**Installation**
      pip install neattext

####**Usage**
+ La méthode OOP (méthode orientée objet)
+ NeatText propose 5 classes principales pour travailler avec des données textuelles
+ **TextFrame : un objet semblable à un cadre pour nettoyer le texte**
+ **TextCleaner : supprimer ou remplacer des spécificités**
+ **TextExtractor : extrayez les données textuelles indésirables**
+ **TextMetrics : statistiques et métriques de mots**
+ **TextPipeline : combiner plusieurs fonctions dans un pipeline**

####**Utilisation de TextFrame**

+ Conserve le texte comme TextFrameobjet.

+ Cela nous permet de faire plus avec notre texte.

+ Il hérite des avantages de TextCleaner et de TextMetrics prêts à l'emploi avec quelques fonctionnalités supplémentaires pour la gestion des données textuelles.

+ C'est le moyen le plus simple pour le prétraitement de texte avec cette bibliothèque, vous pouvez également utiliser les autres classes.

```python
mytext = "This is the mail example@gmail.com ,our WEBSITE is https://example.com 😊."
docx = nt.TextFrame(text=mytext)
docx.text
```

####**2. Pycld2**
+ Et si nous voulons connaître la langue détectée de toutes les phrases ?

+ Dans ce cas, utilisez le code ci-dessous. Vous devez installer la bibliothèque Python pycld2 pour que le code ci-dessous fonctionne.

```python
import pycld2 as cld2

isReliable, textBytesFound, details = cld2.detect(
    "а неправильный формат идентификатора дн назад"
)

print(isReliable)
```

```python
fr_en_Latn = """\
France is the largest country in Western Europe and the third-largest in Europe as a whole.
A accès aux chiens et aux frontaux qui lui ont été il peut consulter et modifier ses collections
et exporter Cet article concerne le pays européen aujourd’hui appelé République française.
Pour d’autres usages du nom France, Pour une aide rapide et effective, veuiller trouver votre aide
dans le menu ci-dessus.
Motoring events began soon after the construction of the first successful gasoline-fueled automobiles.
The quick brown fox jumped over the lazy dog."""

isReliable, textBytesFound, details, vectors = cld2.detect(
    fr_en_Latn, returnVectors=True
)
print(vectors)
```
